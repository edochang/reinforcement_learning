# Non-Stationary 10-Armed Bandit Visualization

This notebook runs the bandit experiment comparing **sample-average** vs **constant step-size** methods and visualizes the results.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import sys
sys.path.insert(0, '..')
from lib.run_bandit import run_bandit
```

## Run the Experiment

Adjust the parameters below to change the experiment configuration.

```python
# Configuration
NUM_SIMULATIONS = 300
NUM_ITERATIONS = 10000
K = 10

print(f"Running {NUM_SIMULATIONS} simulations with {NUM_ITERATIONS} iterations and {K} arms...")
results = run_bandit(
    num_simulations=NUM_SIMULATIONS,
    num_iterations=NUM_ITERATIONS,
    k=K
)
```

## Results

```python
sample_average = {
    'average_rs': results[0],
    'average_best_action_taken': results[1],
}
constant = {
    'average_rs': results[2],
    'average_best_action_taken': results[3],
}

fig, axes = plt.subplots(2, 1, figsize=(10, 8))

# Average Reward
axes[0].plot(sample_average['average_rs'], label='Sample Average')
axes[0].plot(constant['average_rs'], label='Constant Step-Size')
axes[0].set_ylabel('Average Reward')
axes[0].set_title('Average Reward Over Time')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# Ratio of Optimal Actions
axes[1].plot(sample_average['average_best_action_taken'], label='Sample Average')
axes[1].plot(constant['average_best_action_taken'], label='Constant Step-Size')
axes[1].set_ylim([0, 1])
axes[1].set_xlabel('Time Steps')
axes[1].set_ylabel('Ratio of Optimal Actions')
axes[1].set_title('Ratio of Optimal Actions Over Time')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
